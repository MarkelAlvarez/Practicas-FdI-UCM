```
from math import sqrt
num_samples = 100000

def pi_number(n):
  rdd1 = sc.parallelize(range(1, n + 1))
  # Map
  rdd2 = rdd1.map(lambda x: 1/x**2)
  # Reduce
  rdd3 = rdd2.reduce(lambda a, b: a + b)
  pi_result = sqrt(6*rdd3)
  return pi_result

print(pi_number(num_samples))
```
